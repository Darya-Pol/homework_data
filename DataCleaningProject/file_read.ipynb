```python
import pandas as pd
import json
from pprint import pprint

with open('recipes.json') as f:
    recipes = json.load(f)
unique_ingredients=set()
for meal in recipes:
    ingredients = meal['ingredients']
    for ingredient in ingredients:
        unique_ingredients.add(ingredient)
recipes = pd.read_json('recipes.json')
unique_ingredients=list(unique_ingredients)
def contains(ingredient_list):
    if ingredient in ingredient_list:
        return 1
    else:
        return 0
for ingredient in unique_ingredients:
    recipes[ingredient] = recipes['ingredients'].apply(contains)
recipes['ingredients'] = recipes['ingredients'].apply(len)
recipes.to_csv('recipes.csv', index = False)
display(recipes.head())
```
