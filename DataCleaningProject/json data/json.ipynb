```python
import pandas as pd
import json
from pprint import pprint
recipes = pd.read_csv('recipes.csv')
ids = []
id = list(recipes['id'])
for i in id:
    if i not in ids:
        ids.append(i)
ingredients = list(recipes.columns)
ingredients.remove('cuisine')
ingredients.remove('id')
ingredients.remove('ingredients')
new_recipes=[]
def make_list(row):
    list_ingredients = []
    for ingredient in ingredients:
        if row.loc[ingredient] != 0:
            list_ingredients.append(ingredient)
    return list_ingredients
for current_id in ids:
    cuisine = recipes[recipes['id'] == current_id]['cuisine'].iloc[0]
    current_ingredients = make_list(recipes[recipes['id'] == current_id].iloc[0])
    current_recipe = {'cuisine': cuisine, 'id': int(current_id), 'ingredients': current_ingredients}
    new_recipes.append(current_recipe)
new_recipes = json.dumps(new_recipes)
with open("data/new_recipes.json", "w") as write_file:
    write_file.write(new_recipes)
display(new_recipes)
```
